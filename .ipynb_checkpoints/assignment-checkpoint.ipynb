```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset from a CSV file
df = pd.read_csv('D:/PLP AUGUST COHORT/PYTHON PROGRAMMING/Assignment 7/sales_data.csv')

# Displaying the first few rows of the dataset
df.head()
```

```python
# Checking the structure and types of the dataset
df.info()
```

```python
# Checking for missing values in the dataset
df.isnull().sum()
```

```python
# Generating summary statistics for the numerical columns
df.describe()
```

```python
# If there are missing values, we can fill them with the mean or drop them
df.fillna(df.mean(), inplace=True)  # Filling missing values with the mean of respective columns

# Alternatively, we can drop rows with missing values
# df.dropna(inplace=True)
```

```python
# Converting the 'Date' column to DateTime format
df['Date'] = pd.to_datetime(df['Date'])

# Checking if the conversion was successful
df['Date'].dtype
```

```python
# Grouping the data by 'Product' and calculating the total units sold and revenue
grouped_data = df.groupby(['Product', 'Region'])[['Units Sold', 'Revenue']].sum()

# Displaying the grouped data
grouped_data
```

```python
# Grouping by 'Date' to get the total 'Units Sold' and 'Revenue' by date
daily_sales = df.groupby('Date')[['Units Sold', 'Revenue']].sum()

# Plotting the daily sales over time
plt.figure(figsize=(12, 6))
plt.plot(daily_sales.index, daily_sales['Revenue'], marker='o', label='Revenue')
plt.plot(daily_sales.index, daily_sales['Units Sold'], marker='x', label='Units Sold')
plt.title('Total Revenue and Units Sold Over Time')
plt.xlabel('Date')
plt.ylabel('Total Value')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the distribution of the 'Revenue' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Revenue'], kde=True)
plt.title('Distribution of Revenue')
plt.show()
```

```python
# Plotting sales by 'Product' for both 'Revenue' and 'Units Sold'
plt.figure(figsize=(12, 6))
sns.barplot(x='Product', y='Revenue', data=df, ci=None, palette='Blues')
plt.title('Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()

# Plotting units sold by 'Product'
plt.figure(figsize=(12, 6))
sns.barplot(x='Product', y='Units Sold', data=df, ci=None, palette='Blues')
plt.title('Units Sold by Product')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.show()
```

```python
# Plotting total 'Revenue' by 'Region'
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Revenue', data=df, ci=None, palette='Pastel1')
plt.title('Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.show()

# Plotting total 'Units Sold' by 'Region'
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Units Sold', data=df, ci=None, palette='Pastel1')
plt.title('Units Sold by Region')
plt.xlabel('Region')
plt.ylabel('Units Sold')
plt.show()
```

```python
# Plotting revenue over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Revenue', data=df, marker='o')
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate Profit Margin for each product
df['Profit Margin'] = df['Revenue'] - df['Units Sold']  # Assuming Profit is Revenue minus Costs

# Grouping by Product and calculating average profit margin
profit_margin = df.groupby('Product')['Profit Margin'].mean()

# Plotting profit margin by product
plt.figure(figsize=(12, 6))
sns.barplot(x=profit_margin.index, y=profit_margin.values, palette='coolwarm')
plt.title('Profit Margin by Product')
plt.xlabel('Product')
plt.ylabel('Profit Margin')
plt.show()
```
